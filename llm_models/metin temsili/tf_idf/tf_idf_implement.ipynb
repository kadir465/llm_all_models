```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents=[
    "köpek çok tatlı bir hayvandır",
    "köpek ve kuşlar çok sevimlidir",
    "köpekler sadık hayvanlardır",
    "inkeler süt verirler"
]
```

```python
tfidf_vectorizer=TfidfVectorizer()
```

```python
X= tfidf_vectorizer.fit_transform(documents)
```

```python
feature_names=tfidf_vectorizer.get_feature_names_out()
print(f" Kelime Kümesi: {feature_names}")
```

```python
vektor_temsili=X.toarray()
print(f"Vektör Temsili:\n {vektor_temsili}")
```

```python
df_tfidf=pd.DataFrame(vektor_temsili, columns=feature_names)
print(f"TF-IDF DataFrame:\n {df_tfidf}")
```

```python
tf_idf=df_tfidf.mean(axis=0)
print(f"Ortalama TF-IDF Değerleri:\n {tf_idf}")

#yüksek olan sık geçen kelimeler
```
