```python
import nltk
from nltk.tag import hmm 
```

```python
train_data=[
    [('I', 'PRP'), ('love', 'VBP'), ('natural', 'JJ'), ('language', 'NN'), ('processing', 'NN')],
    [('Natural', 'JJ'), ('language', 'NN'), ('processing', 'NN'), ('is', 'VBZ'), ('fascinating', 'JJ')],
    [('I', 'PRP'), ('enjoy', 'VBP'), ('learning', 'VBG'), ('about', 'IN'), ('n-gram', 'NN'), ('models', 'NNS')],
    [('I', 'PRP'), ('love', 'VBP'), ('machine', 'NN'), ('learning', 'NN')],
    [('n-gram', 'NN'), ('models', 'NNS'), ('are', 'VBP'), ('useful', 'JJ'), ('for', 'IN'), ('text', 'NN'), ('prediction', 'NN')]
]
```

```python
trainer=hmm.HiddenMarkovModelTrainer()
hmm_mdoel=trainer.train_supervised(train_data)
```

```python
test_sentence="I love natural language processing".split()
tagged_sentence=hmm_mdoel.tag(test_sentence)
print(tagged_sentence)
```
