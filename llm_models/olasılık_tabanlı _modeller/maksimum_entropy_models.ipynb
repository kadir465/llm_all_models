```python
from nltk.classify import MaxentClassifier
```

```python
train_data=[
    ({'word': 'good'}, 'positive'),
    ({'word': 'bad'}, 'negative'),
    ({'word': 'happy'}, 'positive'),
    ({'word': 'sad'}, 'negative'),
    ({'word': 'excellent'}, 'positive'),
    ({'word': 'terrible'}, 'negative'),
    ({'word': 'amazing'}, 'positive'),
    ({'word': 'awful'}, 'negative'),
    ({'word': 'love'}, 'positive'),
    ({'word': 'hate'}, 'negative')
]
```

```python
classifer=MaxentClassifier.train(train_data, max_iter=10)
```

```python
test_sentence="I love natural language processing"
features={word: (word in test_sentence.lower()) for word in ['good', 'bad', 'happy', 'sad', 'excellent', 'terrible', 'amazing', 'awful', 'love', 'hate']}
```

```python
label=classifer.classify(features)
print(f"The sentiment of the sentence is: {label}")
```
