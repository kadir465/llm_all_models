```python
import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

from collections import Counter
```

```python
corpus=[
    "I love you natural language processing",
    "Natural language processing is fascinating",
    "I love you learning about n-gram models",
    "I love machine learning",
    "N-gram models are useful for text prediction"
]
```

```python
# bir kelimeden sonra gelecek olan kelimeleri tahmin etmek için bigram modeli kullanabiliriz. Örneğin, "I love" ifadesinden sonra hangi kelimenin gelme olasılığını hesaplayabiliriz.

tokens=[word_tokenize(sentence.lower()) for sentence in corpus]
print(tokens)
```

```python
bigrams = []
for token_list in tokens:
    bigrams.extend(list(ngrams(token_list, 2)))

bigrams_freq = Counter(bigrams)
print(bigrams_freq.most_common(10))
```

```python
trigrams = []
for token_list in tokens:
    trigrams.extend(list(ngrams(token_list, 3)))
trigrams_freq = Counter(trigrams)
print(trigrams_freq.most_common(10))
```

```python
bigram=('I', 'love')
prob_you = trigrams_freq[tuple(word.lower() for word in bigram) + ('you',)] / bigrams_freq[tuple(word.lower() for word in bigram)]
print(f"P(you | {bigram}) = {prob_you:.4f}")
```

I + love ' den sonra you gelme olasılığı 0.6 yani bu kelimeleri aldıktan sonra model cümle kuracaksa bu tahmin oranına göre getirebilir
örneğin: dataset büyük olsa ve olasılık değeri 0.8 den yüksek olan kelimleri yazdırsak anlamlı cümleler tamamlayabilir
