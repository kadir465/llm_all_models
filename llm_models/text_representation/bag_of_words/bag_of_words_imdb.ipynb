```python
def clean_text(text):
    # büyük küçük harf dönüşümü
    text=text.lower()
    # rakamların temizlenmesi
    text=re.sub(r"\d+","",text)
    # noktalama işaretlerinin temizlenmesi
    text=re.sub(r"\W\s","",text)
    # kısa kelimelerin temizlenmesi
    cleaned_doc=[word for word in cleaned_doc if word not in stop_words and len(word)>2]
    return text

cleaned_doc=[clean_text(doc) for doc in documents]
```

```python
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(cleaned_doc[:75])
```

```python
feature_names=vectorizer.get_feature_names_out()
print(f"feature names: {feature_names}")
```

```python
words_counts=X.sum(axis=0).A1
word_freq=dict(zip(feature_names,words_counts))
print("kelime frekansı:\n",word_freq)
```
